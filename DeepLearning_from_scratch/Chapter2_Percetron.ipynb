# Perceptron

***Frank Rosenblatt*** invented the concept of the perceptron. Scientists had discovered that brain cells (Neurons) receive input from our senses by electrical signals. The Neurons, then again, use electrical signals to store information, and to make decisions based on previous input. He had the idea that Pecptrons could simulate brain principles, with the ability to learn and make decisions. [Original Paper](https://blogs.umass.edu/brain-wars/files/2016/03/rosenblatt-1957.pdf)


Perceptrons devised from neurons take multiple signals as input and outputs as a single signal. The perecptron signal can have two values, flowing and not flowing, and we will assume that 1 means flowing and 2 means not flowing.


image.png

This is an example of a perceptron that take 2 input signals. $x_1$ and $x_2$ are input signals, $y$ is output signal, and $w_1$ and $w_2$ mean weight. We also call that circle as 'Neurons' or a 'Nodes'. The perceptron will only output 1 when sum of signals crosses the threshold.

$$
y = \left\{
\begin{aligned}
& 0 &  (w_1x_1 + w_2x_2 \leq \theta) \\
& 1 & (w_1x_1 + w_2x_2 > \theta)
\end{aligned}
\right.
$$



# Simple logic circuits



## AND GATE with perceptrons

image.png

````python
def AND(x1, x2):
  w1, w2, theta = 0.5, 0.5, 0.7

  tmp = x1 * w1 + x2 * w2

  if tmp <= theta:
    return 0
  elif tmp > theta:
    return 1
````

````python
print(AND(0, 0))
print(AND(1, 0))
print(AND(0, 1))
print(AND(1, 1))
````

If we replace $\theta$ in the above formula with $-b$, we get the formula below

$$
y = \left\{
\begin{aligned}
& 0 &  (b + w_1x_1 + w_2x_2 \leq 0) \\
& 1 & (b + w_1x_1 + w_2x_2 > 0)
\end{aligned}
\right.
$$



*   b = bias
*   w = weight

The key point: $w_n$ is the parameter that controls how much each input signal affects the outcome, and *bias* is the the paramether that controls how easily the neuron activates (outputs 1 as the outcome).  


````python
import numpy as np
x = np.array([0, 1])
w = np.array([0.5, 0.5])
b = -0.7
````

````python
print(w * x)
print(np.sum(w*x))
print(np.sum(w*x) + b)
````

````python
def AND2(x1, x2):
  x = np.array([x1, x2])
  w = np.array([0.5, 0.5])
  b = -0.7

  tmp = np.sum(w*x) + b

  if tmp <= 0:
    return 0
  else:
    return 1
````

````python
print(AND2(0, 0))
print(AND2(1, 0))
print(AND2(0, 1))
print(AND2(1, 1))
````

## NAND GATE and OR GATE with perceptrons

````python
def NAND(x1, x2):
  x = np.array([x1, x2])
  w = np.array([-0.5, -0.5])
  b = 0.7

  tmp = np.sum(w*x) + b

  if tmp <= 0:
    return 0
  else:
    return 1
````

````python
print(NAND(0, 0))
print(NAND(1, 0))
print(NAND(0, 1))
print(NAND(1, 1))
````

````python
def OR(x1, x2):
  x = np.array([x1, x2])
  w = np.array([0.5, 0.5])
  b = -0.2

  tmp = np.sum(w*x) + b

  if tmp <= 0:
    return 0
  else:
    return 1
````

````python
print(OR(0, 0))
print(OR(1, 0))
print(OR(0, 1))
print(OR(1, 1))
````

# Multi-Layer Perceptron

We can implement an XOR GATE by using the AND and OR GATEs we previously implemented in a *two-layer structure*.

````python
def XOR(x1, x2):
  s1 = NAND(x1, x2)
  s2 = OR(x1, x2)
  y = AND(s1, s2)
  return y
````

````python
print(XOR(0, 0))
print(XOR(1, 0))
print(XOR(0, 1))
print(XOR(1, 1))
````

Like this, multi=layer perceptron can implement things that one-layer percept cannot do that.
